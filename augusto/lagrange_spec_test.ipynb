```julia
include("../HPFEM_mod/HPFEM_mod.jl")
include("../src/HPFEM.jl")
using PyPlot
```

```julia
fun(x) = sin(2*pi*x)

EMM = Float64[lagrange_spec(M,M,1,fun) for M = 3:50]
EMM2 = Float64[lagrange_spec(M,M+2,1,fun) for M = 3:50]
EMe = Float64[lagrange_spec(4,4,nel,fun) for nel = 3:50]
interval = Float64[3:50]
loglog(interval,EMM,label = "Q = M")
loglog(interval,EMM2,label = "Q = M +2 ")
loglog(interval,EMe,label = "M=4,Q=4 e nel = [3:50]")
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
