```julia
fun(x) = sin(2*pi*x)
Marray = 3:50
j = 2

erros = Float64[lagrange_spec(M,M,1,fun) for M in Marray]
Marray2 = Float64[Marray;]
loglog(Marray2,erros,"r",label="Q = M")

for j in 2
    erros2  = Float64[lagrange_spec(M,M+j,1,fun) for M in Marray]
    loglog(Marray2,erros2,label = "Q = M+$j")
end    
legend()
title("loglog")
```
